```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Define colors
road_color = (50, 50, 50)
white = (255, 255, 255)
car_color = (0, 120, 215)        # Modern blue
obstacle_color = (255, 165, 0)   # Bright orange
text_color = white

# Set display dimensions
width = 600
height = 800
display = pygame.display.set_mode((width, height))
pygame.display.set_caption('ðŸš— Car Dodge Game')

# Set game clock
clock = pygame.time.Clock()
car_width = 50
car_height = 100
car_speed = 10
obstacle_speed = 7

# Load car image (colored rectangle)
car_image = pygame.Surface((car_width, car_height))
car_image.fill(car_color)

# Font setup
font_style = pygame.font.SysFont("segoeui", 35)

def draw_text_centered(text, y, color=text_color):
    message = font_style.render(text, True, color)
    text_rect = message.get_rect(center=(width // 2, y))
    display.blit(message, text_rect)

def draw_obstacle(obstacle_x, obstacle_y):
    pygame.draw.rect(display, obstacle_color, [obstacle_x, obstacle_y, car_width, car_height])

def gameLoop():
    game_over = False
    game_close = False

    car_x = width / 2 - car_width / 2
    car_y = height - car_height - 10

    obstacle_x = random.randrange(0, width - car_width)
    obstacle_y = -600

    while not game_over:

        while game_close:
            display.fill(road_color)
            draw_text_centered("You Crashed!", height // 3)
            draw_text_centered("Press C to Play Again or Q to Quit", height // 3 + 50)
            pygame.display.update()

            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_c:
                        gameLoop()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True

        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT] and car_x > 0:
            car_x -= car_speed
        if keys[pygame.K_RIGHT] and car_x < width - car_width:
            car_x += car_speed

        display.fill(road_color)
        draw_obstacle(obstacle_x, obstacle_y)
        obstacle_y += obstacle_speed

        if obstacle_y > height:
            obstacle_y = -car_height
            obstacle_x = random.randrange(0, width - car_width)

        # Check for collision
        if (car_y < obstacle_y + car_height and
            car_y + car_height > obstacle_y and
            car_x < obstacle_x + car_width and
            car_x + car_width > obstacle_x):
            game_close = True

        display.blit(car_image, (car_x, car_y))
        pygame.display.update()
        clock.tick(60)

    pygame.quit()
    quit()

gameLoop()
```
