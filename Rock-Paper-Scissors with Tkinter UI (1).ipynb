```python
import random

print('Winning rules of the game ROCK PAPER SCISSORS are:\n'
      + "Rock vs Paper -> Paper wins \n"
      + "Rock vs Scissors -> Rock wins \n"
      + "Paper vs Scissors -> Scissors wins \n")

choices = ['rock', 'paper', 'scissors']

while True:
    print("Enter your choice: Rock, Paper, or Scissors")

    user_input = input("Enter your choice: ").lower().strip()

    while user_input not in choices:
        user_input = input("Invalid input. Please enter Rock, Paper, or Scissors: ").lower().strip()

    choice_name = user_input.capitalize()
    print('User choice is:', choice_name)

    print("Now it's Computer's Turn...")
    comp_choice = random.choice(choices)
    comp_choice_name = comp_choice.capitalize()
    print("Computer choice is:", comp_choice_name)

    print(choice_name, 'vs', comp_choice_name)

    if user_input == comp_choice:
        print("<== It's a tie! ==>")
    elif (user_input == 'rock' and comp_choice == 'scissors') or \
         (user_input == 'paper' and comp_choice == 'rock') or \
         (user_input == 'scissors' and comp_choice == 'paper'):
        print("<== User wins! ==>")
    else:
        print("<== Computer wins! ==>")

    ans = input("Do you want to play again? (Y/N): ").lower()
    if ans != 'y':
        break

print("Thanks for playing!")
```
